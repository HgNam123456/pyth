```python
contributors = repo.get_contributors()

# Calculate the total number of contributors
total_contributors = contributors.totalCount

# Initialize a dictionary to store the number of commits per contributor
contributor_commits = {}

# Process each contributor
for contributor in contributors:
    contributor_commits[contributor.login] = contributor.contributions

# Find the top contributor
top_contributor = max(contributor_commits, key=contributor_commits.get)
top_contributor_commits = contributor_commits[top_contributor]

# Calculate the total number of commits in the repository
total_commits = sum(contributor_commits.values())

# Calculate the percentage of commits by the top contributor
top_contributor_percentage = (top_contributor_commits / total_commits) * 100

# Output the results
total_contributors, top_contributor, top_contributor_commits, top_contributor_percentage
```

```python
contributors = repo.get_contributors()

# Initialize a dictionary to store the first and last commit dates for each contributor
contributor_dates = {}

# Process each contributor
for contributor in contributors:
    commits = repo.get_commits(author=contributor.login)
    commit_dates = [commit.commit.author.date for commit in commits]
    if commit_dates:
        first_commit = min(commit_dates)
        last_commit = max(commit_dates)
        contributor_dates[contributor.login] = (first_commit, last_commit)

# Calculate the activity period for each contributor
contributor_periods = {contributor: (dates[1] - dates[0]).days for contributor, dates in contributor_dates.items()}

# Find the contributor with the longest period of activity
longest_active_contributor = max(contributor_periods, key=contributor_periods.get)
longest_active_period = contributor_periods[longest_active_contributor]

# Calculate the number of commits per month for each contributor
contributor_activity = {}
for contributor in contributors:
    commits = repo.get_commits(author=contributor.login)
    commit_dates = [commit.commit.author.date for commit in commits]
    commit_months = [commit_date.strftime("%Y-%m") for commit_date in commit_dates]
    monthly_commits = pd.Series(commit_months).value_counts()
    contributor_activity[contributor.login] = monthly_commits.mean()

# Find the most consistent contributor
most_consistent_contributor = max(contributor_activity, key=contributor_activity.get)
most_consistent_activity = contributor_activity[most_consistent_contributor]

# Output the results
longest_active_contributor, longest_active_period, most_consistent_contributor, most_consistent_activity
```

```python
issues = repo.get_issues(state='all')

# Initialize counters
total_open_issues = 0
total_closed_issues = 0
resolution_times = []

# Process each issue
for issue in issues:
    if issue.state == 'open':
        total_open_issues += 1
    else:
        total_closed_issues += 1
        if issue.closed_at and issue.created_at:
            resolution_time = issue.closed_at - issue.created_at
            resolution_times.append(resolution_time.total_seconds())

# Calculate average resolution time in days
if resolution_times:
    avg_resolution_time_seconds = sum(resolution_times) / len(resolution_times)
    avg_resolution_time_days = avg_resolution_time_seconds / (24 * 3600)
else:
    avg_resolution_time_days = None

# Output the results
total_open_issues, total_closed_issues, avg_resolution_time_days
```

```python
pulls = repo.get_pulls(state='all')

# Initialize counters
total_pull_requests = 0
merged_pull_requests = 0
closed_pull_requests = 0
merged_times = []

# Process each pull request
for pr in pulls:
    total_pull_requests += 1
    if pr.merged:
        merged_pull_requests += 1
        if pr.merged_at and pr.created_at:
            merge_time = pr.merged_at - pr.created_at
            merged_times.append(merge_time.total_seconds())
    elif pr.state == 'closed':
        closed_pull_requests += 1

# Calculate average merge time in days
if merged_times:
    avg_merge_time_seconds = sum(merged_times) / len(merged_times)
    avg_merge_time_days = avg_merge_time_seconds / (24 * 3600)
else:
    avg_merge_time_days = None

# Calculate the acceptance rate
if total_pull_requests > 0:
    acceptance_rate = merged_pull_requests / total_pull_requests
else:
    acceptance_rate = None

# Calculate the rejection rate (closed but not merged)
if total_pull_requests > 0:
    rejection_rate = (closed_pull_requests - merged_pull_requests) / total_pull_requests
else:
    rejection_rate = None

# Output the results
total_pull_requests, merged_pull_requests, avg_merge_time_days, acceptance_rate, rejection_rate, closed_pull_requests
```

```python
# Get the contents of the repository
contents = repo.get_contents("")

# Initialize a list to store file sizes
file_sizes = []

# Recursive function to get all files in the repository
def get_all_files(contents):
    while contents:
        file_content = contents.pop(0)
        if file_content.type == "dir":
            contents.extend(repo.get_contents(file_content.path))
        else:
            file_sizes.append(file_content.size)

# Get all files and their sizes
get_all_files(contents)

# Calculate the average file size
average_file_size = sum(file_sizes) / len(file_sizes) if file_sizes else 0

average_file_size
```

```python

# Get all commits
commits = repo.get_commits()

# Initialize a dictionary to store the number of commits per month
commit_counts = {}
contributor_counts = {}

# Process each commit
for commit in commits:
    commit_date = commit.commit.author.date
    month = commit_date.strftime("%Y-%m")
    if month not in commit_counts:
        commit_counts[month] = 0
        contributor_counts[month] = set()
    commit_counts[month] += 1
    contributor_counts[month].add(commit.commit.author.email)

# Convert to pandas DataFrame
commit_data = pd.DataFrame(list(commit_counts.items()), columns=['Month', 'Commits'])
commit_data['Contributors'] = commit_data['Month'].apply(lambda x: len(contributor_counts[x]))
commit_data['Month'] = pd.to_datetime(commit_data['Month'])
commit_data = commit_data.sort_values('Month')

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(commit_data['Month'], commit_data['Commits'], label='Commits', marker='o')
plt.plot(commit_data['Month'], commit_data['Contributors'], label='Contributors', marker='s')
plt.title('Sự tăng trưởng của repo theo thời gian')
plt.xlabel('Thời gian')
plt.ylabel('Số lượng')
plt.legend()
plt.grid(True)
plt.show()
```

```python
commits = repo.get_commits()

# Initialize a dictionary to store the contributors of each commit
commit_contributors = {}

# Process each commit
for commit in commits:
    commit_sha = commit.sha
    commit_contributors[commit_sha] = commit.commit.author.email

# Create a graph
G = nx.Graph()

# Add edges between contributors who have collaborated on the same commit
for commit in commits:
    author = commit.commit.author.email
    for file in commit.files:
        if file.filename not in commit_contributors:
            commit_contributors[file.filename] = set()
        commit_contributors[file.filename].add(author)

# Add nodes and edges to the graph
for file, contributors in commit_contributors.items():
    contributors = list(contributors)
    for i in range(len(contributors)):
        for j in range(i + 1, len(contributors)):
            G.add_edge(contributors[i], contributors[j])

# Draw the graph
plt.figure(figsize=(20, 20))
pos = nx.spring_layout(G, k=0.3)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold", edge_color="gray")
plt.title('Network graph of contributors collaboration')
plt.show()

```

```python
commit_counts_by_year = {}
contributor_counts_by_year = {}

# Get all commits
commits = repo.get_commits()

# Count commits by year
for commit in commits:
    commit_year = commit.commit.author.date.year
    if commit_year in commit_counts_by_year:
        commit_counts_by_year[commit_year] += 1
    else:
        commit_counts_by_year[commit_year] = 1

# Get all contributors and their contributions
contributors = repo.get_contributors()

# Count contributors by year
for contributor in contributors:
    for commit in repo.get_commits(author=contributor.login):
        commit_year = commit.commit.author.date.year
        if commit_year in contributor_counts_by_year:
            contributor_counts_by_year[commit_year].add(contributor.login)
        else:
            contributor_counts_by_year[commit_year] = {contributor.login}

# Convert the contributor sets to counts
for year in contributor_counts_by_year:
    contributor_counts_by_year[year] = len(contributor_counts_by_year[year])

# Convert the dictionaries to DataFrames for easier plotting
commit_counts_df = pd.DataFrame(list(commit_counts_by_year.items()), columns=['Year', 'Commits'])
contributor_counts_df = pd.DataFrame(list(contributor_counts_by_year.items()), columns=['Year', 'Contributors'])

# Merge the DataFrames
data_df = pd.merge(commit_counts_df, contributor_counts_df, on='Year', how='outer').fillna(0)
data_df = data_df.sort_values('Year')

# Filter for the last 10 years
current_year = datetime.now().year
last_10_years_df = data_df[data_df['Year'] >= current_year - 10]

# Plot the data
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(last_10_years_df['Year'], last_10_years_df['Commits'], color='skyblue', label='Commits')
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Commits', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

ax2 = ax1.twinx()
ax2.plot(last_10_years_df['Year'], last_10_years_df['Contributors'], color='green', marker='o', label='Contributors')
ax2.set_ylabel('Number of Contributors', color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.title(f'Number of Commits and Contributors per Year in the {repo_name} Repository')
fig.tight_layout()
plt.show()
```

```python
commits = repo.get_commits()

# Initialize dictionaries to store contributor data
contributor_data = {}

# Process commits to collect data
for commit in commits:
    author = commit.author.login if commit.author else None
    if author:
        commit_date = commit.commit.author.date
        if author in contributor_data:
            contributor_data[author]['commits'].append(commit_date)
        else:
            contributor_data[author] = {'commits': [commit_date]}

# Prepare data for plotting
contributor_activity = []
contributor_frequency = []

for contributor, data in contributor_data.items():
    first_commit = min(data['commits'])
    last_commit = max(data['commits'])
    num_commits = len(data['commits'])
    contributor_activity.append((contributor, first_commit, last_commit))
    contributor_frequency.append((contributor, num_commits))

# Convert to DataFrame
activity_df = pd.DataFrame(contributor_activity, columns=['Contributor', 'First Commit', 'Last Commit'])
frequency_df = pd.DataFrame(contributor_frequency, columns=['Contributor', 'Number of Commits'])

# Sort DataFrames
activity_df = activity_df.sort_values(by='First Commit')
frequency_df = frequency_df.sort_values(by='Number of Commits', ascending=False)

# Plot activity (Gantt chart)
fig, ax = plt.subplots(figsize=(12, 8))
for i, row in activity_df.iterrows():
    ax.plot([row['First Commit'], row['Last Commit']], [i, i], marker='|', linestyle='-', color='skyblue')
ax.set_yticks(range(len(activity_df)))
ax.set_yticklabels(activity_df['Contributor'])
ax.set_xlabel('Date')
ax.set_title('Contributor Activity Over Time')

# Plot frequency
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(frequency_df['Contributor'], frequency_df['Number of Commits'], color='skyblue')
ax.set_xlabel('Contributor')
ax.set_ylabel('Number of Commits')
ax.set_title('Commit Frequency by Contributor')
ax.tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

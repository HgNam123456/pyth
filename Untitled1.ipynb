```python
commits = repo.get_commits()

# Initialize a dictionary to store the contributors of each commit
commit_contributors = {}

# Process each commit
for commit in commits:
    commit_sha = commit.sha
    commit_contributors[commit_sha] = commit.commit.author.email

# Create a graph
G = nx.Graph()

# Add edges between contributors who have collaborated on the same commit
for commit in commits:
    author = commit.commit.author.email
    for file in commit.files:
        if file.filename not in commit_contributors:
            commit_contributors[file.filename] = set()
        commit_contributors[file.filename].add(author)

# Add nodes and edges to the graph
for file, contributors in commit_contributors.items():
    contributors = list(contributors)
    for i in range(len(contributors)):
        for j in range(i + 1, len(contributors)):
            G.add_edge(contributors[i], contributors[j])

# Draw the graph
plt.figure(figsize=(20, 20))
pos = nx.spring_layout(G, k=0.3)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold", edge_color="gray")
plt.title('Network graph of contributors collaboration')
plt.show()
```

```python
contributors = repo.get_contributors()

# Initialize a list to store contributor data
contributor_data = []

# Get all pull requests
all_pulls = repo.get_pulls(state='all')
pull_request_data = [{'login': pr.user.login, 'number': pr.number} for pr in all_pulls]

# Get all issues
all_issues = repo.get_issues(state='all')
issue_data = [{'login': issue.user.login, 'number': issue.number} for issue in all_issues]

for contributor in contributors:
    commits = repo.get_commits(author=contributor).totalCount
    pull_requests = len([pr for pr in pull_request_data if pr['login'] == contributor.login])
    issues = len([issue for issue in issue_data if issue['login'] == contributor.login])
    contributor_data.append({
        'login': contributor.login,
        'commits': commits,
        'pull_requests': pull_requests,
        'issues': issues
    })

# Create a DataFrame from the data
df_contributors = pd.DataFrame(contributor_data)

# Calculate basic statistics
total_commits = df_contributors['commits'].sum()
total_pull_requests = df_contributors['pull_requests'].sum()
total_issues = df_contributors['issues'].sum()

# Print statistics
print(f"Total commits: {total_commits}")
print(f"Total pull requests: {total_pull_requests}")
print(f"Total issues: {total_issues}")

# Print contributor data
print(df_contributors)
# Plot the number of commits, pull requests, and issues by contributor
plt.figure(figsize=(14, 8))
df_contributors.set_index('login')[['commits', 'pull_requests', 'issues']].plot(kind='bar')
plt.title('Commits, Pull Requests, and Issues by Contributor')
plt.xlabel('Contributor')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
